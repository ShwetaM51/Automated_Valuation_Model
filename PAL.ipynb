```python
import re
def county(s):
    rx01 = (re.search(r'\w+(?=\s+\d$)', s) or None,)[0] #followed by a number (dublin 6)
    rx1 = (re.search(r'\w+(?=\s+\d\d$)', s) or None,)[0] #followed by a number (dublin 16)
    rx2 = (re.search(r'(?:co.|co|county)\s\w+$',s) or None,)[0]#Prefix as co.
    rx3 = (re.search(r'(?:(?<![\w\d])\bco.\b(?![\w\d])|(?<![\w\d])\bco\b(?![\w\d])|(?<![\w\d])\bcounty\b(?![\w\d])) (\w+)',s) or None,)[0]
    #rx3 = (re.search(r'(?:\bco.\b|\bco\b|\bcounty\b) (\w+)',s) or None,)[0]
    rx4 = (re.search(r'\w+(?=\s\d+[\,$]\s\b([a-z]+[0-9]+|[0-9]+[a-z]+)[a-z0-9]*\b)', s) or None,)[0]
    
    rx5 = (re.search(r'\w+(?=\s+city\s+centre+$)', s) or None,)[0] #followed by city centre
    rx6 = (re.search(r'\w+(?=\s+city+$)', s) or None,)[0] #followed by city
    
    rx7 = (re.search(r'\w+(?= \b([a-z]+[0-9]+|[0-9]+[a-z]+)[a-z0-9]*\b[\,$]\s\b([a-z]+[0-9]+|[0-9]+[a-z]+)[a-z0-9]*\b)', s) or None,)[0] #dublin d16, d08 yx5e
    rx8 = (re.search(r'\w+(?=[\,$]\s\b([a-z]+[0-9]+|[0-9]+[a-z]+)[a-z0-9]*\b)', s) or None,)[0] #dublin, d08 yx5e
    
    rx9=  (re.search(r'\w+(?=\s+\d+[\,$]\s+north)', s) or None,)[0] #dublin 18, north dublin city
    rx91=  (re.search(r'\w+(?=\s+\d+[\,$]\s+south)', s) or None,)[0] #dublin 18, south dublin city
    rx10=(re.search(r'(?:east|west) (\w+)$',s) or None) #east cork
    rx11 = (re.search(r'\w+(?=[a-zA-Z0-9 ]*$)', s) or None,)[0] #followed alphanumeric with space (Eirocde)
    rx12 = (re.search(r'\w+(?=\s+\w+$)', s) or None,)[0] #followed alphanumeric with space (Eirocde)
    rx13 = (re.search(r'\b(\w+)$', s) or None,)[0]    #simple last word
    

    if rx01:
        return rx01.group(),
    elif rx1:
        return rx1.group()
    elif rx2:
        r=rx2.group()
        if 'co.' in r:
            return re.sub('co. ','',r) 
        elif 'co' in r:
            return re.sub('co ','',r) 
        else:
            return re.sub('county ','',r) 
    elif rx3: 
        return rx3.group(1)
    elif rx4:
        return rx4.group(0)
    elif rx5:
        return rx5.group()
    elif rx6:
        return rx6.group()
    elif rx7:
        return rx7.group()
    elif rx8:
        return rx8.group()
    elif rx9:
        return rx9.group()
    elif rx91:
        return rx91.group()
    elif rx10:
        return rx10.group(1)
    elif rx11:
        return rx11.group()
    elif rx12:
        return rx12.group()
    elif rx13:
        return rx13.group()
    else:
        return 'No county'
    
```
